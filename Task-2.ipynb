```python
# f.	Create a table with the number of referrals each month.
```

```python
data.info()
```

```python
data.groupby([data.Created_at.dt.year, data.Created_at.dt.month]).count().rename_axis(['year', 'month'])['Created_at'].reset_index(name='No: of referrals')
```

```python
F_DataFrame = data.groupby([data.Created_at.dt.year, data.Created_at.dt.month]).count().rename_axis(['Year', 'Month'])['Created_at'].reset_index(name='No: of referrals')
```
